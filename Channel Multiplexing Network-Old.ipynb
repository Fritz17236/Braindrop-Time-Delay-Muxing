
# Channel Multiplexing for Signal Input

The goal here is to compress multiple signal dimensions into 1 signal as input to a neural network.  




```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import nengo
from numpy import sin,cos,pi
```

```python

#Parameters
t_syn = .01 # Synaptic Time Constant for Neuron Pool
t_sig = .01 # Period of Input Signal Multiplexing, effectively the inverse sampling rate of each multiplexed signal
dt    = .001
```

# First let's implement a simple time multiplexor:

```python
####################################################################################################################################
######## Multiplex Function ########################################################################################################
# Given a time window, split the window evenly between each dimension of the input channel
#(time multiplexing of multiple channels)
def chan_mux(t,u,t_sig=t_sig):
    u = np.asarray(u)
    D = u.size   # Number of Dimensions in Signal 

    # Decide which signal (i) will pass through, depending on the time
    for i in np.arange(D):
        if i*t_sig/D <= np.mod(t,t_sig) and np.mod(t,t_sig) < (i+1)*t_sig/D:
            idx = i
    # Quick way to make a unit vector
    e = np.zeros((D,1))
    e[idx] = 1
    return np.dot(np.squeeze(e),u)
###################################################################################################################################
###################################################################################################################################

###################################################################################################################################
###################################################################################################################################
#For Testing Mux Output
dt    = .001
z = np.zeros((4,100))
y = np.zeros((1,100))

for idx in np.arange(1,100):
    z[0,idx] = 1#  np.sin(.5*ts)
    z[1,idx] = 2#  .2*np.sin(.6*t)
    z[2,idx] = 3#  .3*np.sin(.8*t)
    z[3,idx] = 4#  .4*np.sin(1*t)
    
    y[0,idx] = chan_mux(dt*idx,z[:,idx],t_sig)
 
plt.plot(np.arange(y.size),z[0,:],label='Signal 1')
plt.plot(np.arange(y.size),z[1,:],label='Signal 2')
plt.plot(np.arange(y.size),z[2,:],label='Signal 3')
plt.plot(np.arange(y.size),z[3,:],label='Signal 4')
plt.plot(np.arange(y.size),np.squeeze(y),label='Muxed Signal')
plt.title('Time-Multiplexed Input Signal')
plt.legend()
plt.show()
```

```python
###################################################################################################################################
################# Now Try it On Nengo-Generated Gaussian Noise ####################################################################

# Create 4 Channels of Gaussian White Noise
with nengo.Network() as model:
    a = nengo.Node(nengo.processes.WhiteSignal(1.0, high=5, seed=0))
    b = nengo.Node(nengo.processes.WhiteSignal(1.0, high=10, seed=0))
    c = nengo.Node(nengo.processes.WhiteSignal(1.0, high=5, rms=0.3, seed=0))
    d = nengo.Node(nengo.processes.WhiteSignal(0.5, high=5, seed=0))
    ap = nengo.Probe(a)
    bp = nengo.Probe(b)
    cp = nengo.Probe(c)
    dp = nengo.Probe(d)

# Run simulation to generate signals
with nengo.Simulator(model) as sim:
    sim.run(1.0)
```

```python
# Now multiplex the signal: 

u = np.squeeze(np.asarray([
    sim.data[ap],
    sim.data[bp],
    sim.data[cp],
    sim.data[dp],
]))

u_mux = []
idx   = 0
for t in sim.trange():
    u_mux += [chan_mux(t,u[:,idx],t_sig)]
    idx += 1

plt.plot(sim.trange(),u_mux,label='Multiplexed Signal')
plt.plot(sim.trange(), u[0,:], label='5 Hz cutoff')
plt.plot(sim.trange(), u[1,:], label='10 Hz cutoff')
plt.plot(sim.trange(), u[2,:], label='5 Hz cutoff, 0.3 RMS amplitude')
plt.plot(sim.trange(), u[3,:], label='5 Hz cutoff, 0.5 s period')
plt.xlabel("time [s]")
plt.legend(loc=2)
plt.show()
```

# Now lets drive a nengo ensemble with a multiplexed input process

```python
# Define a simple mux that only takes time as an argument. 
# This makes it much easier to interface as a nengo process,
# since we don't have to worry about how nengo passes inputs
# and parameters. 

t_sig =.01

def simple_mux(t):
    u = np.asarray([sin(2*pi*t),cos(-t)]) # Hardcoded Mux Input Channels

    D = u.size   # Number of Dimensions in Signal
    
    

    # Decide which signal (i) will pass through, depending on the time
    for i in np.arange(D):
        if i*t_sig/D <= np.mod(t,t_sig) and np.mod(t,t_sig) < (i+1)*t_sig/D:
            idx = i
            
    # Quick way to make a unit vector
    e = np.zeros((D,1))
    e[idx] = 1
    
    return np.dot(np.squeeze(e),u)

def ideal_mult(w):
    assert w.size == 2
     # Receives input window of 2 points: input(t) & input(t-t_sig) where t_sig is the mux sampling rate.
     # Computes their product and returns
    
    return w[0]*w[1]


# Now let's turn this simple mux into a process
class Mux(nengo.Process):
    def make_step(self, shape_in, shape_out, dt, rng):
        
        state = np.array([0.0])

        # define the step function, which will be called
        # by the node every time step
        def step(t):
            state[:] =  simple_mux(t)
            return state

        return step  # return the step function


```

# We'll try to compute an arbitrary function of two inputs (addition) that are multiplexed. 

```python


with nengo.Network() as model:
    
    # Create Input Node
    mux_input = nengo.Node(Mux(), size_in=0, size_out=1)
    A = nengo.Ensemble(1000,dimensions=1)
    B = nengo.Ensemble(1000,dimensions=1)   
    
    syn_val = .01
    nengo.Connection(mux_input,A,synapse=syn_val)
    nengo.Connection(A,B,function=np.mean)
    
    Ap = nengo.Probe(A)
    Bp = nengo.Probe(B,synapse=syn_val)
    muxp = nengo.Probe(mux_input)


with nengo.Simulator(model) as sim:
    sim.run(2)

diff = (ideal_func(sim.trange())-sim.data[Ap])
rmse =  np.sqrt(np.mean((diff)**2))
    
plt.figure()

plt.plot(sim.trange(), sim.data[muxp],label='Input')
plt.plot(sim.trange(), sim.data[Bp],label='Output')
plt.plot(sim.trange(),ideal_func(sim.trange()),label='Ideal Output')
plt.legend()
plt.xlabel('time [s]');
print("RMSE: %f" % rmse)
```
